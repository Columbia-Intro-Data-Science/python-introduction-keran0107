````python
# Code here
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2)
x_train= train[['TV','radio','newspaper']]
y_train= train[['sales']]
x_test= test[['TV','radio','newspaper']]
y_test= test[['sales']]
````

d) Train model on training data, and make predictions on testing data

````python
import numpy as np
# Code here
regr = LinearRegression()
regr.fit(x_train, y_train)
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(x_test) - y_test) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x_test, y_test))
````

e) Evalute the R^2 on **testing** data. Is this good? Bad? Why?

````python
SSreg = np.mean((regr.predict(x_test) - y_test) ** 2)
SStot = np.mean((y_test - np.mean(y_test)) ** 2)
print(1-SSreg/SStot)
#R^2 is equal to 0.92, we can say that this model works good for the test data.
````

## How could you have improved performance?

*Hint:* Try plotting the data in three dimensions along with the hyperplane solution, and see if you can infer
a new variable which will help, or try a nonlinear/non-parametric model

````python
# Type your answer here. 
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.formula.api as sm
from matplotlib import cm

csv = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
model = sm.ols(formula='sales ~ TV + radio', data = csv)
fit = model.fit()

fit.summary()

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

x_surf = np.arange(0, 350, 20)                # generate a mesh
y_surf = np.arange(0, 60, 4)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

exog = pd.core.frame.DataFrame({'TV': x_surf.ravel(), 'radio': y_surf.ravel()})
out = fit.predict(exog = exog)
ax.plot_surface(x_surf, y_surf,
                out.reshape(x_surf.shape),
                rstride=1,
                cstride=1,
                color='None',
                alpha = 0.4,cmap=cm.coolwarm)

ax.scatter(csv['TV'], csv['radio'], csv['sales'],
           c='blue',
           marker='o',
           alpha=1)

ax.set_xlabel('TV')
ax.set_ylabel('radio')
ax.set_zlabel('sales')

plt.show()
````

````python
# We can try to add the interative to decrease the predict error
df['TV*radio']=df['TV']*df['radio']
````

````python
train, test = train_test_split(df, test_size=0.2)
y_train= train['sales']
y_test=test['sales']
X_train=train[['TV','radio','TV*radio','newspaper']]
X_test=test[['TV','radio','TV*radio','newspaper']]
````

````python
regr = LinearRegression()
regr.fit(X_train, y_train)
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(X_test) - y_test) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(X_test, y_test))
````

````python
print('R^2:',regr.score(X_test,y_test))
````

````python
plt.title('R^2')
plt.scatter(regr.predict(X_test),y_test)
plt.plot(y_test,y_test)
````

## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python

from numpy import *
# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations):
    points = genfromtxt("data/data.csv", delimiter=",")
    learning_rate = 0.0001
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print ("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1],color='b')
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')

run(100)
````

````python
# Use this to visually discuss convergence rate based on learning rate

for num in range(0,10):
    run(num)
    plt.show()
````

For this method, we update our slop and intercept every time by using learning rate and gradient calculated by all the points. The more times we run the iteration, we can find the our model will get covergence to the points in the plot. 
new_b = b_current - (learningRate * b_gradient)
new_m = m_current - (learningRate * m_gradient)
From the equation above, we can see that every new b and new m can be updated with the learning rate. The larger the learning rate, the more change of b and m. So we can get the covergence more quickly with a larger learning rate. But if the learning rate is too large, the best fit model might be missed with few times iteration. Also, if the learning rate is too small, we should run the iteration many times to find the good fit, which takes longer time.

Plot the error as a function of the number of iterations for various learning rates. Choose the rates

````python
def run_learningrate(num_iterations, learning_rate):
    points = genfromtxt("data/data.csv", delimiter=",")
    learning_rate = learning_rate
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print ("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1],color='b')
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')

for num in range(0,10):       
 for a in [0.00005,0.0001,0.0002,0.0003,0.0004,0.0005,0.001,0.002]:
    run_learningrate(num,learning_rate=a)
    plt.show()
````

Here I chose the learning rate from 0.00005 to 0.002 and number of iteration from 0 t 10, we can easily see that the for the low number of iterations, model can sometimes have a better fit with a slightly larges learning_rate. The performance of small learning rates can be increased with the increasing number of iterations. But for the large learning rates, it can never have a good performance.
